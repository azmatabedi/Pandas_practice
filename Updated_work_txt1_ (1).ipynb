#### 7a. There are too many columns in this DataFrame, we want a new DataFrame with less columns.  Combine `Room.Board`, `Books`, `Personal` into 1 column `Full cost` in the original DF.  Use `sum` instead of `+`

```python
#your code here
columns_to_sum = ['Room.Board','Books','Personal']
df['Full Cost']= df[columns_to_sum].sum(axis=1)
df
```

#### 7b. Add a new column `% part time` that calculates % of part time undergraduates out of total undergraduates (see legend above to decide how to calculate this) with integer percents using `apply` with values `high` (for >20%), `medium` (between 10 and 20% including), `low` (for <10%). 

```python
#your code here
def func(x, y=0):
      if x > 20 and x > y:
          return 'high'
      if x>10 and x<20 and x>y:
          return 'medium'
      else:
          return 'low'

df['% part time'] = (df['P.Undergrad']/df['F.Undergrad'])*100
df['% part time'] = df['% part time'].apply(func)
df
```

#### 7c. Would you say there are more colleges with < 10% or > 20% part time students? 

```python
#your code here
val = df['% part time'].value_counts(ascending=True)
if val.high > val.low:
  print('more colleges with >20% part time students')
else:
  print('more colleges with >10% part time students')
```

#### 8a. By what percent is the maximum `Full cost` greater in private colleges vs. public colleges?

```python
#your code here
# df1 = df['Private']
# df1
val = df[df['Private']=='No']
val
val1 = df[df['Private']=='Yes']
val1
p1 = val['Full Cost'].sum()
p2 = val1['Full Cost'].sum()
p1 = (p1/df['Full Cost'].sum())*100
p2 = (p2/df['Full Cost'].sum())*100
print(p2-p1)
```

#### 8b. By what percent is the average `Full cost` greater in private colleges vs. public colleges?

```python
#your code here
avg1 = val['Full Cost'].mean()
avg2 = val1['Full Cost'].mean()
avg1 = (avg1/df['Full Cost'].mean())*100
avg2 = (avg2/df['Full Cost'].mean())*100
print(avg2-avg1)
```
